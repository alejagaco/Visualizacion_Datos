```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos= pd.read_csv("/content/drive/MyDrive/Coding Dojo/Data Science_Fundamentals/4_Semana/Median Age at First Marriage .csv")
datos.head()
```

```python
datos.describe()
```

```python
sns.lmplot(data=datos, x='Median Age', y='Population Density per Sq Mi',hue='TimeFrame');
plt.legend(loc='upper left')
plt.title('Edad media matrimonio por densidad de población', fontsize=12);
```

```python
plt.figure(dpi=150)
sns.lineplot(data=datos, x='Name',y='Median Age', hue='TimeFrame');
plt.xticks(rotation=90);
plt.xlabel('State',fontsize=8)
plt.ylabel('Median Age',fontsize=8)
plt.tick_params(labelsize=5) # Para cambiar el tamaño de los valores del eje x

plt.legend(loc='lower left',fontsize=7)
plt.tight_layout()
```
